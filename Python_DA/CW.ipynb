```python
Матрица goods описывает количество муки, яиц и сахара в трех видах десертов некоторой кондитерской. 
Известно, что затраты на выпечку первого десерта составляют 2 евро, второго – 4 евро, третьего – 3 евро. 
Найдите, сколько должны стоить мука, яйца и сахар для такой себестоимости десертов.
# Расход продуктов
goods = np.array(
[[0.5, 6, 0.1],
[1, 10, 0.5],
[0.3, 10, 0.6]]
)
Для решения задачи составить систему уравнений, найти определитель этой системы. Если определитель не равен нулю, найти обратную матрицу и значения переменных.
2 = 0.5*x + 6*y + 0.1*z
4 = 1*x + 10*y + 0.5*z 
3 = 0.3*x + 10*y + 0.6*z
Где:

x - стоимость муки
y - стоимость яиц
z - стоимость сахара
```

```python
import numpy as np

# Матрица коэффициентов (затраты муки, яиц, сахара)
A = np.array([[0.5, 6, 0.1],
              [1, 10, 0.5],
              [0.3, 10, 0.6]])


# Вектор себестоимости десертов
b = np.array([2, 4, 3])

# Вычисляем определитель матрицы A
det_A = np.linalg.det(A)
print(f"Определитель матрицы A: {det_A:.2f}")

# Если определитель не равен 0, решаем систему
if det_A != 0:
    A_inv = np.linalg.inv(A)  # Обратная матрица
    X = np.dot(A_inv, b)      # Решение системы (стоимость муки, яиц, сахара)
    print("Определитель матрицы A:", det_A)
    print("Стоимость муки, яиц и сахара:", X)
else:
    print("Матрица A вырожденная, решение невозможно.")
```

```python
import numpy as np

# Матрица расхода продуктов
goods = np.array(([[0.5, 6, 0.1],
                   [1, 10, 0.5],
                   [0.3, 10, 0.6]])

# Затраты на выпечку десертов
costs = np.array([2, 4, 3])

# Вычисляем определитель матрицы goods
det_goods = np.linalg.det(goods)
print(f"Определитель матрицы goods: {det_goods:.2f}")

# Если определитель не равен нулю, решаем систему уравнений
if det_goods != 0:
    # Находим обратную матрицу
    goods_inv = np.linalg.inv(goods)
    
    # Решаем систему уравнений
    solution = goods_inv @ costs
    
    print("Решение системы уравнений:")
    print(f"Стоимость муки: {solution[0]:.2f} евро")
    print(f"Стоимость яиц: {solution[1]:.2f} евро") 
    print(f"Стоимость сахара: {solution[2]:.2f} евро")
else:
    print("Определитель матрицы равен нулю, система уравнений не имеет единственного решения.")
```

```python
# Фирма закупает ноутбуки и стационарные компьютеры для оборудования офисов на двух этажах
# здания.
# ● На первом этаже планируется поставить 20 стационарных компьютеров и 20 ноутбуков.
# ● На втором этаже – 12 стационарных компьютеров и 35 ноутбуков.
# Компьютер стоит 850 евро, ноутбук – 1125 евро.
# 1. Запишите матрицу А, описывающую количество компьютеров и ноутбуков на каждом
# этаже.
# 2. Запишите вектор-столбец с ценами компьютеров и ноутбуков.
# 3. С помощью матричных операций вычислите стоимость оборудования на каждом этаже.
```

```python
import numpy as np

# Матрица количества компьютеров и ноутбуков на этажах
A = np.array([[20, 20], 
              [12, 35]])

# Вектор цен компьютеров и ноутбуков
prices = np.array([850, 1125])

# Вычисляем стоимость оборудования на каждом этаже
costs = np.dot(A, prices)

print("Стоимость оборудования на каждом этаже:")
print(f"Первый этаж: {costs[0]:.2f} евро")
print(f"Второй этаж: {costs[1]:.2f} евро")
```
