```python
prices = np.random.randint(1000,1300,7) 

price_c = np.diff(prices)
print(prices)
print("изменение цены за каждый день:", price_c)

days_of_week = ["понедельник", "вторник", "среда", "четверг", "пятница", "суббота", "воскресенье"]

max = np.argmax(price_c)
min = np.argmin(price_c)

good_day = days_of_week[max + 1]  # +1, т.к. первое значение не с чем сравнивать
bad_day = days_of_week[min + 1]

print("Лучший день:", good_day)
print("Худший день:", bad_day)
```

```python
4. Пять спортсменов участвовали в соревновании, в котором нужно было бежать 2 часа, плыть
1 час и ехать на велосипеде 5 часов.
Первый спортсмен бежал со скоростью 15 км/ч, плыл со скоростью 4 км/ч и ехал на
велосипеде со скоростью 25 км/ч.
Соответствующие скорости второго спортсмена: 14 км/ч, 5 км/ч, 27 км/ч.
Третий спортсмен: 16 км/ч, 5 км/ч, 23 км/ч.
Четвертый: 13 км/ч, 5.5 км/ч, 26 км/ч.
Пятый: 12 км/ч, 4.5 км/ч, 30 км/ч.
Побеждает тот, кто преодолел большее расстояние. Напишите программу, которая
определяет победителя.
(Подсказка: Подумайте, что здесь можно записать матрицей, что вектором, какая матричная
операция определяет пройденное расстояние.)
```

```python
t = np.array([2, 1, 5])  # Бег, плавание, велоспорт

# Скорости спортсменов (в км/ч)
s = np.array([
    [15, 4, 25],  # Первый спортсмен
    [14, 5, 27],  # Второй спортсмен
    [16, 5, 23],  # Третий спортсмен
    [13, 5.5, 26],  # Четвертый спортсмен
    [12, 4.5, 30]   # Пятый спортсмен
])

#  расстояние
dist = s @ t 
print("Расстояния, пройденные спортсменами:", dist)

# победитель
win_i = np.argmax(dist)
win_dist = dist[win_i]
print("Победитель: Спортсмен", win_i + 1, " преодолел:", win_dist, "км")
```
